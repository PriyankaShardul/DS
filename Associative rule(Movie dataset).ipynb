```python
#Import libraries
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_csv(r"C:\DS Assignments\Assignment 9(Association Rules)\movies.csv")
df
```

## Data Preprocessing

```python
# As the data is categorical format, we are using One Hot Encoding to convert into numerical format.
df1=pd.get_dummies(df)
df1
```

```python
df1.shape
```

## Apriori Algorithm

```python
frequent_itemsets=apriori(df1,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
# Association Rule mining
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
```

#### A leverage value of 0 indicates independence. Range will be [-1 1]¶
#### A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions.
rules[rules.lift>1]
```

```python
rules[rules.confidence>=1]
```
